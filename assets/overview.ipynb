```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Gaussian function
def gaussian(x, c, alpha):
    return c * np.exp(-alpha * x ** 2)

# Data for s and p orbitals
s_orbitals = [
    (3047.52488, 0.001834737132),
    (457.369518, 0.01403732281),
    (103.948685, 0.06884262226),
    (29.2101553, 0.2321844432),
    (9.28666296, 0.4679413484),
    (3.16392696, 0.3623119853)
]

sp_orbitals = [
    (78.6827235, -0.01193324198, 0.06899906659),
    (18.8128854, -0.01608541517, 0.3164239610),
    (5.44249258, 0.1143456438, 0.7443082909),
    (1.687144782, 1.0, 1.0),
    (0.438, 1.0, 1.0)
]

# Range of x values
x = np.linspace(-1, 1, 400)

# Plotting the s orbitals
plt.figure(figsize=(12, 6))
for alpha, c in s_orbitals:
    y = gaussian(x, c, alpha)
    plt.plot(x, y, label=f'alpha={alpha}, c={c}')

plt.title('Gaussian Type Orbital Basis - s Orbitals for Carbon')
plt.xlabel('x')
plt.ylabel('ψ(x)')
plt.legend()
plt.grid(True)
plt.show()

# Plotting the sp orbitals
plt.figure(figsize=(12, 6))
for alpha, c_s, c_p in sp_orbitals:
    y_s = gaussian(x, c_s, alpha)
    y_p = gaussian(x, c_p, alpha)
    plt.plot(x, y_s, label=f's: alpha={alpha}, c={c_s}')
    plt.plot(x, y_p, '--', label=f'p: alpha={alpha}, c={c_p}')

plt.title('Gaussian Type Orbital Basis - sp Orbitals for Carbon')
plt.xlabel('x')
plt.ylabel('ψ(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import genlaguerre, factorial

# Constants
a0 = 1  # Bohr radius (arbitrary units)

# Radial wave function
def radial_wave_function(n, l, r):
    rho = 2 * r / n
    normalization = np.sqrt((2/n)**3 * factorial(n-l-1) / (2*n*factorial(n+l)))
    laguerre = genlaguerre(n-l-1, 2*l+1)
    return normalization * (rho**l) * np.exp(-rho/2) * laguerre(rho)

# Range of r values
r = np.linspace(0, 20, 400)

# Plotting
plt.figure(figsize=(4, 3))

# 1s orbital
plt.plot(r, radial_wave_function(1, 0, r), label='1s', color='tab:blue')

# 2s orbital
plt.plot(r, radial_wave_function(2, 0, r), label='2s', color='tab:orange')

# 2p orbital (l=1)
plt.plot(r, radial_wave_function(2, 1, r), label='2p', color='tab:orange', linestyle='--')

# 3s orbital
plt.plot(r, radial_wave_function(3, 0, r), label='3s', color='tab:green')

# 3p orbital (l=1)
plt.plot(r, radial_wave_function(3, 1, r), label='3p', color='tab:green', linestyle='--')

# 3d orbital (l=2)
plt.plot(r, radial_wave_function(3, 2, r), label='3d', color='tab:green', linestyle=':')


# plt.title('Radial Parts of Spherical Harmonics for Hydrogen-like Atom')
# plt.xlabel('r')
plt.ylabel('GTO(r)')
plt.xlim(0, 10)
plt.ylim(-0.1, 0.8)
plt.yticks([])
plt.xticks([])
# plt xticks label
plt.text(0, -0.12, 'r', ha='center', va='top')
plt.legend(frameon=False)
plt.grid(False)
# spines
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.show()
```
