```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import ticker
from matplotlib.ticker import ScalarFormatter
from scipy.optimize import curve_fit

def func(x, a, b):
    return a * x * np.log(x) + b


goodred = "#DF6B63"
goodblue = "#546BDF"
goodpurple = "#8C6CE7"
goodgreen = "#40DE66"
goodpink = "#FF9999"

fig, ax = plt.subplots()
fig.set_size_inches(5.5, 2.5)
fig.set_dpi(600)
ax.set_facecolor('white')

x = np.array([
    10**3, # 10
    20**3, # 20
    25**3, # 25
    30**3, # 30
    35**3, # 35
    40**3, # 40
    50**3, # 50
    ])
y_train = np.array([
    3.2742676734924316, # 10
    4.865520000457764, # 20
    5.918131113052368, # 30
    9.122117519378662, # 35
    11.10208511352539, # 40
])
y_inf = np.array([
    2.6540205478668213, # 10
    4.147275447845459, # 20
    5.208123207092285, # 30
    7.876112461090088, # 35
    9.988749265670776, # 40
])



popt1, pcov1 = curve_fit(func, x, y_train)
popt2, pcov2 = curve_fit(func, x, y_inf)


# coef1 = np.polyfit(x,y_inf,1)
# poly1d_fn1 = np.poly1d(coef1)
# coef2 = np.polyfit(x,y_train,1)
# poly1d_fn2 = np.poly1d(coef2)

plt.plot(x,y_train, 'o', color=goodgreen,label="Train time")
plt.plot(x,func(x,popt1[0], popt1[1]), '--',color=goodgreen,alpha=0.5)
plt.plot(x,y_inf, 'o',color=goodpink, label="Inference time")
plt.plot(x,func(x,popt2[0], popt2[1]), '--',color=goodpink,alpha=0.5)


ax.tick_params(direction='in', which='both')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
ax.grid(True,which='major',linestyle='--',linewidth=0.5)
plt.xlim(0,65000)
plt.legend()
plt.xlabel("Number of probe nodes $(M)$")
plt.ylabel(f"Time (s)")
plt.show()
```
