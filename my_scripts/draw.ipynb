```python
import torch
```

```python
import matplotlib.pyplot as plt
```

```python
x = torch.linspace(0,3*torch.pi,100)
y = torch.sin(x)
size = (10,2)
linewidth=4
plt.figure(figsize=size)
plt.plot(x, y, 'k-', label='sin(x)',linewidth=linewidth)

# Remove frame
plt.box(False)

# Remove ticks
plt.xticks([])
plt.yticks([])
```
